```python
import random

NUM_ROUNDS = 5  # Number of rounds to play

def main():
    print("🎉 Welcome to the High-Low Game! 🎉")
    print('--------------------------------')

    # Track the score
    your_score = 0

    # Play multiple rounds
    for i in range(NUM_ROUNDS):
        print(f"\n🌀 Round {i + 1} 🌀")
        # Generate random numbers for the computer and the player
        computer_num = random.randint(1, 100)
        your_num = random.randint(1, 100)
        print(f"Your number is {your_num} 🔢")

        # Get user input for their choice
        choice = input("🤔 Do you think your number is higher or lower than the computer's? (Enter 'higher' or 'lower'): ").lower()

        # Make sure the player inputs a valid choice (either 'higher' or 'lower')
        while choice not in ["higher", "lower"]:
            choice = input("❌ Invalid input! Please enter 'higher' or 'lower': ").lower()

        # Check if the guess is correct
        higher_and_correct = choice == "higher" and your_num > computer_num
        lower_and_correct = choice == "lower" and your_num < computer_num

        if higher_and_correct or lower_and_correct:
            print(f"🎉 You were right! The computer's number was {computer_num} 🎉")
            your_score += 1  # Increment score for correct guess
        else:
            print(f"😢 Aww, that's incorrect. The computer's number was {computer_num} 😢")

        # Show the player's current score
        print(f"Your score is now {your_score} 🏆")
        print("===================================")

    # Display final score and personalized feedback
    print(f"\n🏁 Game Over! 🏁")
    print(f"🎯 Your final score is {your_score} ⭐")

    if your_score == NUM_ROUNDS:
        print("🌟 Wow! You played perfectly! 🌟")
    elif your_score > NUM_ROUNDS // 2:
        print("🎉 Good job, you played really well! 🎉")
    else:
        print("😅 Better luck next time! Keep practicing! 😅")

    print("Thanks for playing! 👏👏")

# Run the game
if __name__ == "__main__":
    main()
```
