```python
import random

NUM_ROUNDS = 5  # Number of rounds to play

def main():
    print("ğŸ‰ Welcome to the High-Low Game! ğŸ‰")
    print('--------------------------------')

    # Track the score
    your_score = 0

    # Play multiple rounds
    for i in range(NUM_ROUNDS):
        print(f"\nğŸŒ€ Round {i + 1} ğŸŒ€")
        # Generate random numbers for the computer and the player
        computer_num = random.randint(1, 100)
        your_num = random.randint(1, 100)
        print(f"Your number is {your_num} ğŸ”¢")

        # Get user input for their choice
        choice = input("ğŸ¤” Do you think your number is higher or lower than the computer's? (Enter 'higher' or 'lower'): ").lower()

        # Make sure the player inputs a valid choice (either 'higher' or 'lower')
        while choice not in ["higher", "lower"]:
            choice = input("âŒ Invalid input! Please enter 'higher' or 'lower': ").lower()

        # Check if the guess is correct
        higher_and_correct = choice == "higher" and your_num > computer_num
        lower_and_correct = choice == "lower" and your_num < computer_num

        if higher_and_correct or lower_and_correct:
            print(f"ğŸ‰ You were right! The computer's number was {computer_num} ğŸ‰")
            your_score += 1  # Increment score for correct guess
        else:
            print(f"ğŸ˜¢ Aww, that's incorrect. The computer's number was {computer_num} ğŸ˜¢")

        # Show the player's current score
        print(f"Your score is now {your_score} ğŸ†")
        print("===================================")

    # Display final score and personalized feedback
    print(f"\nğŸ Game Over! ğŸ")
    print(f"ğŸ¯ Your final score is {your_score} â­")

    if your_score == NUM_ROUNDS:
        print("ğŸŒŸ Wow! You played perfectly! ğŸŒŸ")
    elif your_score > NUM_ROUNDS // 2:
        print("ğŸ‰ Good job, you played really well! ğŸ‰")
    else:
        print("ğŸ˜… Better luck next time! Keep practicing! ğŸ˜…")

    print("Thanks for playing! ğŸ‘ğŸ‘")

# Run the game
if __name__ == "__main__":
    main()
```
