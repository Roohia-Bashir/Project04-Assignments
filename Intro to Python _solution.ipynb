```python
# Gravitational constants for various planets in the solar system
MERCURY_GRAVITY = 0.376
VENUS_GRAVITY = 0.889
MARS_GRAVITY = 0.378
JUPITER_GRAVITY = 2.36
SATURN_GRAVITY = 1.081
URANUS_GRAVITY = 0.815
NEPTUNE_GRAVITY = 1.14
EARTH_GRAVITY = 1.0

def main():
    # Step 1: Prompt the user for their weight on Earth
    earth_weight = float(input("Enter a weight on Earth: "))

    # Step 2: Prompt the user for the name of a planet
    planet = input("Enter a planet: ").capitalize()  # Ensure the first letter is capitalized

    # Step 3: Determine the gravitational constant for the selected planet
    if planet == "Mercury":
        gravity_constant = MERCURY_GRAVITY
    elif planet == "Venus":
        gravity_constant = VENUS_GRAVITY
    elif planet == "Mars":
        gravity_constant = MARS_GRAVITY
    elif planet == "Jupiter":
        gravity_constant = JUPITER_GRAVITY
    elif planet == "Saturn":
        gravity_constant = SATURN_GRAVITY
    elif planet == "Uranus":
        gravity_constant = URANUS_GRAVITY
    elif planet == "Neptune":
        gravity_constant = NEPTUNE_GRAVITY
    else:
        gravity_constant = EARTH_GRAVITY  # Default to Earth if no match

    # Step 4: Calculate the equivalent weight on the selected planet
    planetary_weight = earth_weight * gravity_constant

    # Step 5: Round the result to two decimal places
    rounded_planetary_weight = round(planetary_weight, 2)

    # Step 6: Print the result
    print(f"The equivalent weight on {planet}: {rounded_planetary_weight}")

# Run the program only if it is the main script being executed
if __name__ == "__main__":
    main()
```
