```python
import random

# Function to start the Guess the Number Game
def guess_the_number():
    # Generate a random number between 1 and 20
    number_to_guess = random.randint(1, 20)

    # Initialize variables
    guesses_taken = 0
    guess = None

    print("Welcome to Guess the Number Game!")
    print("I'm thinking of a number between 1 and 20.")

    # While loop to keep asking for guesses until the correct one is made
    while guess != number_to_guess:
        try:
            # Get the user's guess
            guess = int(input("Take a guess: "))

            # Check if the guess is within the valid range
            if guess < 1 or guess > 20:
                print("Please guess a number between 1 and 20.")
                continue

            guesses_taken += 1

            # Check if the guess is too low, too high, or correct
            if guess < number_to_guess:
                print("Too low! Try again.")
            elif guess > number_to_guess:
                print("Too high! Try again.")
            else:
                print(f"Congratulations! You guessed the number in {guesses_taken} attempts.")
                break  # Exit the loop when the number is guessed correctly
        except ValueError:
            print("Invalid input. Please enter a valid integer.")

# Start the game
guess_the_number()
```
