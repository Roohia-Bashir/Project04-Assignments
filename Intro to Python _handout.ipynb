```python
# Planetary Weight Calculator

def calculate_weight_on_planet(weight, planet):
    # Planet gravity constants (as percentages of Earth's gravity)
    gravity_constants = {
        "Mercury": 0.376,
        "Venus": 0.889,
        "Mars": 0.378,
        "Jupiter": 2.360,
        "Saturn": 1.081,
        "Uranus": 0.815,
        "Neptune": 1.140,
        "Earth": 1.000  # Earth gravity is 100%
    }

    # Capitalize the first letter to handle user input like 'earth' or 'EARTH'
    planet = planet.capitalize()

    # Check if the planet exists in the dictionary
    if planet in gravity_constants:
        # If Earth is chosen, no calculation needed (as gravity constant is 1.0)
        if planet == "Earth":
            return weight
        # Calculate weight on the given planet
        planet_weight = weight * gravity_constants[planet]
        # Round the result to 2 decimal places
        return round(planet_weight, 2)
    else:
        return "Invalid planet name."

def main():
    # Get user's weight on Earth
    weight_on_earth = float(input("Enter a weight on Earth: "))

    # Get user's selected planet
    planet = input("Enter a planet: ")

    # Calculate the weight on the selected planet
    weight_on_planet = calculate_weight_on_planet(weight_on_earth, planet)

    # Print the result
    if isinstance(weight_on_planet, float):
        print(f"The equivalent weight on {planet}: {weight_on_planet}")
    else:
        print(weight_on_planet)  # Invalid planet name

# Run the program
if __name__ == "__main__":
    main()
```
