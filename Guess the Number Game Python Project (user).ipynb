```python
import random

def guess_the_fruit():
    fruits = ["Apple 🍎", "Mango 🥭", "Orange 🍊"]
    colors = {"Apple 🍎": "Red", "Mango 🥭": "Yellow", "Orange 🍊": "Orange"}
    chosen_fruit = random.choice(fruits)
    correct_color = colors[chosen_fruit]

    print("Welcome to the Guess the Fruit Game!")
    print("I'm thinking of a fruit: Apple 🍎, Mango 🥭, or Orange 🍊.")
    print("Can you guess which fruit I am thinking of by choosing the correct color?")

    # Ask the player for the color
    guess_color = input(f"Which color matches the fruit I'm thinking of? (Red, Yellow, or Orange): ").capitalize()

    # Check if the player's guess is correct
    if guess_color == correct_color:
        print(f"Yay! You guessed it! The fruit was {chosen_fruit} with color {correct_color}! 🎉")
    else:
        print(f"Oops! The correct fruit was {chosen_fruit} with color {correct_color}. Try again next time!")

guess_the_fruit()
```
