```python
def main():
    # Prompt the user for their weight on Earth
    weight_on_earth = float(input("Enter your weight on Earth: "))

    # Prompt the user for the name of a planet
    planet = input("Enter the name of the planet: ").capitalize()  # Capitalize the planet name to handle cases like 'earth' or 'EARTH'

    # Calculate and print the equivalent weight on the selected planet
    if planet == "Mars":
        weight_on_mars = weight_on_earth * 0.38
        print(f"Your equivalent weight on Mars is: {weight_on_mars:.2f}")
    elif planet == "Jupiter":
        weight_on_jupiter = weight_on_earth * 2.34
        print(f"Your equivalent weight on Jupiter is: {weight_on_jupiter:.2f}")
    elif planet == "Venus":
        weight_on_venus = weight_on_earth * 0.91
        print(f"Your equivalent weight on Venus is: {weight_on_venus:.2f}")
    elif planet == "Mercury":
        weight_on_mercury = weight_on_earth * 0.38
        print(f"Your equivalent weight on Mercury is: {weight_on_mercury:.2f}")
    elif planet == "Saturn":
        weight_on_saturn = weight_on_earth * 1.06
        print(f"Your equivalent weight on Saturn is: {weight_on_saturn:.2f}")
    elif planet == "Uranus":
        weight_on_uranus = weight_on_earth * 0.92
        print(f"Your equivalent weight on Uranus is: {weight_on_uranus:.2f}")
    elif planet == "Neptune":
        weight_on_neptune = weight_on_earth * 1.19
        print(f"Your equivalent weight on Neptune is: {weight_on_neptune:.2f}")
    elif planet == "Earth":
        print(f"Your weight on Earth is: {weight_on_earth:.2f}")
    else:
        print("Invalid planet name entered. Please enter a valid planet name.")

if __name__ == "__main__":
    main()
```
