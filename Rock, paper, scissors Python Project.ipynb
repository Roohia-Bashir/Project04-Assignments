```python
import random

def play():
    # Ask the user for their choice
    user_choice = input("Choose Rock (r), Paper (p), or Scissors (s): ").lower()

    # If the user enters something wrong, ask again
    if user_choice not in ['r', 'p', 's']:
        return "Invalid choice! Please pick 'r' for Rock, 'p' for Paper, or 's' for Scissors."

    # Let the computer pick its choice
    computer_choice = random.choice(['r', 'p', 's'])

    # Compare choices and return the result
    if user_choice == computer_choice:
        return "It's a tie!"

    if (user_choice == 'r' and computer_choice == 's') or \
       (user_choice == 's' and computer_choice == 'p') or \
       (user_choice == 'p' and computer_choice == 'r'):
        return "You win!"

    return "You lose!"

# Run the game
print(play())
```
