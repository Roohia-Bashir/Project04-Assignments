```python
#Index Game Solution:

def access_element(lst, index):
    try:
        return lst[index]
    except IndexError:
        return "âš ï¸ Index out of range. Try again!"

def modify_element(lst, index, new_value):
    try:
        lst[index] = new_value
        return f"âœ… List updated: {lst}"
    except IndexError:
        return "âš ï¸ Index out of range. Please choose a valid index!"

def slice_list(lst, start, end):
    try:
        return lst[start:end]
    except IndexError:
        return "âŒ Invalid indices. Check the range!"

def index_game():
    lst = [1, 2, 3, 4, 5]  # Example list
    print("ğŸ® Welcome to the Index Game!")
    print("Current list: ğŸ²", lst)
    print("Choose an operation: 'access' to view, 'modify' to change, 'slice' to extract parts.")

    operation = input("Enter your choice: ").lower()

    if operation == "access":
        index = int(input("ğŸ” Enter the index you want to access: "))
        print(access_element(lst, index))
    elif operation == "modify":
        index = int(input("ğŸ› ï¸ Enter the index to modify: "))
        new_value = input("âœï¸ Enter the new value: ")
        print(modify_element(lst, index, new_value))
    elif operation == "slice":
        start = int(input("ğŸ§© Enter the start index: "))
        end = int(input("ğŸ§© Enter the end index: "))
        print(slice_list(lst, start, end))
    else:
        print("ğŸš« Invalid operation. Please choose 'access', 'modify', or 'slice'.")

# Start the game ğŸ®
index_game()
```
