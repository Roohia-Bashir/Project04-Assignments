```python
#Index Game Solution:

def access_element(lst, index):
    try:
        return lst[index]
    except IndexError:
        return "⚠️ Index out of range. Try again!"

def modify_element(lst, index, new_value):
    try:
        lst[index] = new_value
        return f"✅ List updated: {lst}"
    except IndexError:
        return "⚠️ Index out of range. Please choose a valid index!"

def slice_list(lst, start, end):
    try:
        return lst[start:end]
    except IndexError:
        return "❌ Invalid indices. Check the range!"

def index_game():
    lst = [1, 2, 3, 4, 5]  # Example list
    print("🎮 Welcome to the Index Game!")
    print("Current list: 🎲", lst)
    print("Choose an operation: 'access' to view, 'modify' to change, 'slice' to extract parts.")

    operation = input("Enter your choice: ").lower()

    if operation == "access":
        index = int(input("🔍 Enter the index you want to access: "))
        print(access_element(lst, index))
    elif operation == "modify":
        index = int(input("🛠️ Enter the index to modify: "))
        new_value = input("✏️ Enter the new value: ")
        print(modify_element(lst, index, new_value))
    elif operation == "slice":
        start = int(input("🧩 Enter the start index: "))
        end = int(input("🧩 Enter the end index: "))
        print(slice_list(lst, start, end))
    else:
        print("🚫 Invalid operation. Please choose 'access', 'modify', or 'slice'.")

# Start the game 🎮
index_game()
```
