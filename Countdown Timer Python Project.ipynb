```python
import time

def countdown(t):
    while t:
        mins, secs = divmod(t, 60)
        print(f"{mins:02}:{secs:02} SEC", end="\r")
        time.sleep(1)
        t -= 1

    print("üí• Fire in the hole! üí•")

# Input time in seconds
seconds = int(input("‚è∞ Enter the countdown time in seconds: "))

# Function call
countdown(seconds)
```
