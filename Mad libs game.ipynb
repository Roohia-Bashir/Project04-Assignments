```python
# Mad Libs Game

# Getting user inputs
adjective1 = input("Enter an adjective: ")
adjective2 = input("Enter another adjective: ")
animal = input("Enter an animal: ")
verb = input("Enter a verb (past tense): ")
place = input("Enter a place: ")
plural_noun = input("Enter a plural noun: ")

# Creating the Mad Lib story
story = f"Once upon a time, a very {adjective1} {animal} went to {place}. It was feeling so {adjective2} and decided to {verb} around. Everyone saw it and thought it was the best {plural_noun} they'd ever seen!"

# Displaying the story
print("\nHere is your Mad Libs story:")
print(story)
```
