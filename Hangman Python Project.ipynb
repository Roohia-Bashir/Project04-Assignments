```python
import random

def hangman():
    word_list = ["python", "javascript", "programming", "computer", "science"]
    word = random.choice(word_list)
    lives = 6
    used_letters = set()
    hint_used = False  # To track if hint has been used

    print("🎉 Welcome to Hangman! 🎉")
    print(f"The word has {len(word)} letters. 🤔")

    while lives > 0:
        display_word = [letter if letter in used_letters else '❓' for letter in word]
        print('Word: ', ' '.join(display_word))
        print(f"Lives: {lives} ❤️, Used letters: {' '.join(used_letters)}")

        # Ask for a hint if lives are low and it hasn't been used yet
        if not hint_used and lives <= 3:
            give_hint = input("Would you like a hint? (y/n): 🌟 ").lower()
            if give_hint == 'y':
                # Reveal a random letter from the word
                hint_letter = random.choice([letter for letter in word if letter not in used_letters])
                print(f"Hint: One of the letters is '{hint_letter}' 🔍")
                hint_used = True

        user_letter = input("Guess a letter: ✍️ ").lower()

        if user_letter in word and user_letter not in used_letters:
            used_letters.add(user_letter)
            if all(letter in used_letters for letter in word):
                print(f"🎉 Congrats! You guessed the word: {word} 🎉")
                break
        elif user_letter not in used_letters:
            used_letters.add(user_letter)
            lives -= 1
            print("❌ Incorrect guess! Try again. ❌")
        else:
            print("❗ You already guessed that letter! Please try again. ❗")

        if lives == 0:
            print(f"💀 You lost! The word was {word}. 💀")

if __name__ == "__main__":
    hangman()

```
