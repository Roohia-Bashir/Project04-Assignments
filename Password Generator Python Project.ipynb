```python
import random

# Function to generate a random password
def generate_password(length, characters):
    password = "".join(random.choice(characters) for _ in range(length))
    return password

# Main function
def main():
    print("🔑 Welcome to the Password Generator! 🔑\n")

    # Get user input for the number of passwords and their length
    num_passwords = int(input("🙋‍♂️ How many passwords would you like to generate? "))

    # Clarified prompt for password length
    password_length = int(input("📏 What is the length of each password (in number of characters)? "))

    # Asking the user what characters to include in the password
    print("\nAvailable characters for password generation:")
    print("1. Lowercase letters (a-z)")
    print("2. Uppercase letters (A-Z)")
    print("3. Numbers (0-9)")
    print("4. Special characters (!@#$%^&*)")

    # User input to select the characters they want in the password
    include_lower = input("Do you want to include lowercase letters (y/n)? ").lower() == 'y'
    include_upper = input("Do you want to include uppercase letters (y/n)? ").lower() == 'y'
    include_numbers = input("Do you want to include numbers (y/n)? ").lower() == 'y'
    include_special = input("Do you want to include special characters (y/n)? ").lower() == 'y'

    # Build the characters string based on user choices
    characters = ""
    if include_lower:
        characters += "abcdefghijklmnopqrstuvwxyz"
    if include_upper:
        characters += "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    if include_numbers:
        characters += "1234567890"
    if include_special:
        characters += "!@#$%^&*()"

    # Generate and display the passwords
    print("\nGenerating your passwords... 🎉\n")
    for i in range(num_passwords):
        password = generate_password(password_length, characters)
        print(f"🔐 Password {i + 1}: {password}")

# Run the program
if __name__ == "__main__":
    main()
```
